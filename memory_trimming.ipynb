ConversationBufferWindowMamory è deprecata:

```python
import warnings
warnings.filterwarnings('ignore')
from langchain.memory import ConversationBufferWindowMemory
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
memory = ConversationBufferWindowMemory(k=3)
conversation = ConversationChain(
    llm=llm,
    memory = memory,
    verbose=True
)
```

```python
domande = [
    "Mi chiamo Alessandro e sono un sviluppatore tutto fare!",
    "Quali linguaggi dovrei imparare?",
    "Quanto tempo ci vuole per diventare un architetto del software ?",
    "Quanto tempo ci vuole per certificarmi nel az-305",
    "Come mi chiamo?"
]

for domanda in domande:
  print(f"Domande: {domanda}")
  risposta = conversation.predict(input=domanda)
  print(f"Risposta: {risposta[:100]}")
  print(f"Messaggi in memoria:{len(memory.chat_memory.messages)}")
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage, trim_messages
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
messages = [
    SystemMessage("Sei un assistent"),
    HumanMessage("Mi chiamo Alessandro e sono un sviluppatore tutto fare!"),
    AIMessage("Ciao Alessandro! Sono un'intelligenza artificiale progettata per aiutare con una vasta gamma di compiti. Sono in grado di fornire informazioni dettagliate su molti argomenti diversi. Se hai bisogno di aiuto con qualcosa, non esitare a chiedere!"),
    HumanMessage("Quali linguaggi dovrei imparare?"),
    AIMessage("Ci sono molti linguaggi di programmazione popolari che potresti considerare di imparare, a seconda dei tuoi interessi e delle tue esigenze. Alcuni dei linguaggi più richiesti nel settore includono Python, Java, JavaScript, C ++ e Ruby. Ognuno di questi ha le proprie caratteristiche e usi specifici, quindi potresti voler fare ulteriori ricerche per determinare quale potrebbe essere il migliore per te. Se hai bisogno di ulteriori informazioni su uno specifico linguaggio di programmazione, sarò felice di aiutarti!"),
    HumanMessage("Quanto tempo ci vuole per diventare un architetto del software ?"),
    AIMessage("l tempo necessario per diventare un architetto del software può variare a seconda del tuo livello di esperienza attuale, della tua dedizione e dell'intensità con cui studi e lavori nel settore. In genere, diventare un architetto del software richiede diversi anni di esperienza pratica nella progettazione e nello sviluppo di software, oltre a una solida comprensione dei principi di progettazione e dell'architettura dei sistemi software. Potresti voler considerare l'ottenimento di una laurea in informatica o un campo correlato, oltre a ottenere certificazioni pertinenti e acquisire esperienza lavorativa nel settore. Con impegno e perseveranza, potresti raggiungere il tuo obiettivo di diventare un architetto del software in un periodo di tempo ragionevole."),
    HumanMessage("Quanto tempo ci vuole per prendere la certificazion az-305?"),
    AIMessage("Mi dispiace, ma al momento non ho informazioni specifiche sulla certificazione AZ-305. Ti consiglio di controllare il sito web ufficiale di Microsoft o di contattare direttamente il team di supporto per ottenere informazioni dettagliate sul processo di certificazione e sui requisiti di tempo. Se hai altre domande o hai bisogno di ulteriori informazioni su un argomento diverso, sarò felice di aiutarti!"),
    HumanMessage("Ti ricordi il mio nome?")

]
trimmed = trim_messages(
    messages,
    max_tokens=100,
    strategy="last",
    include_system=True,
    start_on="human",
    token_counter=llm
)
for msg in trimmed:
    print(f"[{msg.type}]:{msg.content[:100]}")
```
