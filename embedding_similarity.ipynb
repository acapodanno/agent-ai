```python
import numpy as np
import matplotlib.pyplot as plt
from openai import OpenAI
from sklearn.decomposition import PCA

client = OpenAI()

def get_embedding(text: str) -> np.ndarray:
  response = client.embeddings.create(
    input=[text],
    model="text-embedding-3-small"
  )
  return np.array(response.data[0].embedding)
frasi = [
    "Il sole splende alto nel cielo azzurro.",
    "Oggi è una bellissima giornata per fare una passeggiata al parco.",
    "Il caffè italiano è famoso in tutto il mondo per il suo aroma intenso.",
    "Mi piace leggere libri di fantascienza e avventura.",
    "La musica classica mi aiuta a rilassarmi e a concentrarmi."
]
embeddings = { frase: get_embedding(frase) for frase in frasi }

embeddings
```

```python
def cosine_similarity(a:np.ndarray, b:np.ndarray) -> float:
  return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
def euclidean_distance(a:np.ndarray, b:np.ndarray) -> float:
  return np.linalg.norm(a - b)
def dot_product(a:np.ndarray,b: np.ndarray) -> float:
  return np.dot(a, b)

confronti = [
    ("Il cielo è limpido e senza nuvole.",
    "Il tempo è bello e luminoso.",
    "Le stelle brillano nel buio della notte."),

    ("È il momento perfetto per una passeggiata nel verde.",
    "Mi godrò una camminata rilassante all'aperto.",
    "Il tempo ideale per una gita in montagna."),

    ("L'espresso italiano ha un sapore unico.",
    "Il profumo del caffè si diffonde nell'aria.",
    "Il tè verde ha proprietà benefiche."),

    ("Sono appassionato di storie di esplorazione spaziale.",
    "Amo le narrazioni epiche e fantastiche.",
    "Preferisco i romanzi storici e i thriller."),

    ("Le sinfonie di Beethoven sono meravigliose.",
    "Ascoltare Mozart mi mette di buon umore.",
    "La musica rock è energica e coinvolgente.")
]
```

```python
confronti_embeddings_by_original_phrase = {}
for i, original_frase in enumerate(frasi):
    if i < len(confronti):
        comparison_phrases_tuple = confronti[i]
        confronti_embeddings_by_original_phrase[original_frase] = [get_embedding(p) for p in comparison_phrases_tuple]
    else:
        print(f"Warning: No corresponding comparison phrases found for original phrase: '{original_frase}'. Skipping.")
```

```python
for i, (original_frase, original_embedding) in enumerate(embeddings.items()): # `embeddings` comes from yYgeDOa4I9EC
    print(f"\n--- Frase Originale ({i+1}): '{original_frase}' ---")

    if original_frase in confronti_embeddings_by_original_phrase:
        # Get the original comparison phrases tuple to display the text
        current_confronti_tuple = confronti[i] # `confronti` comes from IJKoDN5uKhxZ

        for j, comparison_embedding in enumerate(confronti_embeddings_by_original_phrase[original_frase]):
            comparison_text = current_confronti_tuple[j] # Get the text for display

            # `cosine_similarity`, `euclidean_distance`, `dot_product` come from IJKoDN5uKhxZ
            cos_sim = cosine_similarity(original_embedding, comparison_embedding)
            euc_dist = euclidean_distance(original_embedding, comparison_embedding)
            dot_prod = dot_product(original_embedding, comparison_embedding)

            print(f"  > Frase di Confronto ({j+1}): '{comparison_text}'")
            print(f"    Similarità del Coseno: {cos_sim:.4f}")
            print(f"    Distanza Euclidea:    {euc_dist:.4f}")
            print(f"    Prodotto Scalare:     {dot_prod:.4f}")
    else:
        print(f"  Nessuna frase di confronto specifica trovata per questa frase originale.")
```
